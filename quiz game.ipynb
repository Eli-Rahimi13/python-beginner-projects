```python
score = 0

print ("=" * 30)
print (f"{'🎯 Quiz Game':^30}")
print ("=" * 30)

questions = [{"question": "What is the capital of France?","options": ["London", "Paris", "Berlin", "Rome"], "answer": 2},{"question": "Which planet is closet to the Sun?", "options": ["Venus", "Earth", "Mars", "Mecury"], "answer": 4},{"question": "What is 12 x 12?", "options": ["124", "144", "132", "112"], "answer": 2},{"question": "Which language is used for web styling?", "options": ["Python", "HTML", "CSS", "Java"], "answer": 3},{"question": "How many continents are there?", "options": ["5", "6","7","8"], "answer": 3}]

for index, question in enumerate(questions, 1):
    print (f"\nQ{index}: {question['question']}")
    
    for i, option in enumerate(question['options'], 1):
        print (f"{i}. {option}")

    user_answer = int(input("Answer (1-4): "))
    
    if user_answer == question['answer']:
        print ("✅ Correct! +10 points")
        score += 10
    else:
        print ("❌ Wrong!")

print ("-" * 30)

print (f"'Final Score: {score}/50")
print ("Well Done! 🔆")
    
                                      
        
        
```
