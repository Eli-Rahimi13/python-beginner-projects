```python
title = "Our Menu"
again = ''
cart = []
receipt = "Receipt"
summary = {}
total_general = 0

print ("=" * 30)
print (f"{title:^30}")
print ("=" * 30)

menu = {'Apple': 1.50, 'Banana': 0.75, 'Milk': 2.30, 'Bread': 3.20}

for index, (item, price) in enumerate(menu.items(), 1):
    print (f"{index}. {item:.<20}${price:.2f}")

print ("=" * 30)

while True:
    again = input("Add item (or 'done' to finish): ").lower().capitalize()
    
    if again == "Done":
        break
    elif again in menu:
        cart.append(again)
        print(f"{again} added! ✅")
    else:
        print ("Item not found! ❌")
        
print ("=" * 30)
print (f"{receipt:^30}")
print ("=" * 30)

for item in cart:
    if item in summary:
        summary[item] += 1
    else:
        summary[item] = 1
        
for item, count in summary.items():
    price = menu[item]
    total_general += price * count
    print (f"{item} x{count:.<15}${price * count:.2f}")
    
print ("-" * 30) 
print (f"{'Total:':<20}${total_general:.2f}")

    
```
